The goal is to load the file poiList-ijcai15 and take poi_name,poi_theme,latititude,longitude based on user id obtained from file userVisits-ijcai15

```
import pandas as pd
import os
import itertools

path_places = "/content/data-ijcai15/poiList-ijcai15"
path_user = "/content/data-ijcai15/userVisits-ijcai15"

files_places = ["POI-Pert.csv", "POI-Vien.csv", "POI-Buda.csv", "POI-Glas.csv", "POI-Delh.csv", "POI-Osak.csv", "POI-Toro.csv", "POI-Edin.csv"]
files_user = ["userVisits-Pert-allPOI.csv", "userVisits-Vien-allPOI.csv", "userVisits-Buda-allPOI.csv", "userVisits-Glas.csv", "userVisits-Delh-allPOI.csv", "userVisits-Osak.csv", "userVisits-Toro.csv", "userVisits-Edin.csv"]
  
poi,poi_name = {},{}
for i,j in zip(files_user,files_places):
  df_user = pd.read_csv(os.path.join(path_user,i),sep = ";")
  df_poi = pd.read_csv(os.path.join(path_places,j),sep = ";")
  df = df_user.groupby("userID")
  df.first()
  id = [x[0] for x in [x for x in df.userID]]
  for x in id:
    list_of_poiID = list(set(df.get_group(str(x)).poiID))      
    inserted = df_poi.iloc[[y-1 for y in list_of_poiID],1:]
    x = str(x)
    for j in range(len(inserted)):
      poi_name["poi_name"],poi_name["lat"],poi_name["long"],poi_name["theme"] = inserted.iloc[j,0],inserted.iloc[j,1],inserted.iloc[j,2],inserted.iloc[j,3]
      if str(x) in poi.keys():
        poi[x].append(dict(poi_name))
      else:
        poi[x] = [dict(poi_name)]    
```

```
import json
with open("poi.json","w") as outfile:
  json.dump(poi,outfile)
```
